```python
console = Console()
from pathlib import Path

def list_experiments() -> Dict[int, str]:
    # Pega o cwd, depois pula dois pais e aponta para 'experiments'
    experiments_dir = Path.cwd().parents[1] / "experiments"

    if not experiments_dir.is_dir():
        raise FileNotFoundError(f"{experiments_dir} não existe")

    experiments = sorted(
        [p for p in experiments_dir.iterdir() if p.is_dir() and not p.name.startswith(".")],
        key=lambda p: p.name
    )
    return {i: str(p) for i, p in enumerate(experiments)}

# Exemplo de uso:
exp_dirs = list_experiments()

def display_experiment_details(exp_path: str):
    """Display config.yaml using Rich formatting."""
    config_path = os.path.join(exp_path, "config.yaml")
    if not os.path.exists(config_path):
        console.print(f"[red]Missing config.yaml in {exp_path}[/red]")
        return

    with open(config_path, "r") as f:
        config_text = f.read()

    syntax = Syntax(config_text, "yaml", theme="ansi_light", line_numbers=True)
    console.print(Panel(syntax, title=f"[bold green]Experiment: {os.path.basename(exp_path)}"))


def load_organized_results(exp_path: str) -> Dict[str, Dict[str, pd.DataFrame]]:
    """Rebuild organized_results from Excel, with correct sheet names."""
    excel_path = os.path.join(exp_path, "results.xlsx")
    if not os.path.exists(excel_path):
        console.print(f"[red]Missing results.xlsx in {exp_path}[/red]")
        return {}

    xls = pd.ExcelFile(excel_path)
    organized = {}
    for sheet in xls.sheet_names:
        if "__" in sheet:
            variant, metric = sheet.split("__", 1)
        else:
            variant, metric = "default", sheet
        if variant not in organized:
            organized[variant] = {}
        organized[variant][metric] = xls.parse(sheet)
    return organized


def load_all_experiments(base_dir: str = "experiments") -> Dict[str, Dict[str, Dict[str, pd.DataFrame]]]:
    """Load and reconstruct all experiments with cleaned sheet names."""
    experiment_dirs = list_experiments()
    all_results = {}

    for _, exp_path in experiment_dirs.items():
        display_experiment_details(exp_path)
        organized = load_organized_results(exp_path)
        exp_name = os.path.basename(exp_path)
        if organized:
            all_results[exp_name] = organized

    console.print(f"\n[bold green]Loaded {len(all_results)} experiments.[/bold green]")
    return all_results


# Load all experiments without user input
all_organized_results = load_all_experiments()
```

```python
from typing import Callable, Dict, Optional
import pandas as pd
import logging

# Configure logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# --- Transformation Functions ---
def phrase_case(value: str) -> str:
    return value.replace('_', ' ').replace('-', ' ').title()

def upper_case(value: str) -> str:
    return value.upper()

def lower_case(value: str) -> str:
    return value.lower()

def sentence_case(value: str) -> str:
    text = value.replace('_', ' ').replace('-', ' ')
    return text.capitalize()

def capitalize_case(value: str) -> str:
    return value.replace('_', ' ').replace('-', ' ').title()

def replace_symbol(value: str) -> str:
    return value.replace('_', ' ').replace('-', ' ')

# Registry of supported transformations
TRANSFORM_REGISTRY: Dict[str, Callable[[str], str]] = {
    'Phrase': phrase_case,
    'Upper': upper_case,
    'Lower': lower_case,
    'Sentence': sentence_case,
    'Capitalize': capitalize_case,
    'ReplaceSymbol': replace_symbol,
}

import re
import pandas as pd
import numpy as np

def apply_string_transformation(value, transform: str):
    """
    Aplica a transformação especificada em `transform` ao valor `value`.
    Se o valor não for string (ou for NaN), retorna-o inalterado.
    transform pode ser: 'Upper', 'Lower', 'Capitalize', 'Phrase'.
    """
    # Trata valores nulos
    if pd.isna(value):
        return value

    s = str(value)

    t = transform.strip().lower()
    if t == 'upper':
        return s.upper()
    elif t == 'lower':
        return s.lower()
    elif t == 'capitalize':
        # primeira letra maiúscula, resto minúsculas
        return s.capitalize()
    elif t == 'phrase':
        # substitui _ ou - por espaço e capitaliza cada palavra
        # Ex: "minha_coluna-teste" -> "Minha Coluna Teste"
        s = re.sub(r'[_\-]+', ' ', s)
        return ' '.join(w.capitalize() for w in s.split())
    else:
        raise ValueError(f"Transformação desconhecida: {transform!r}")


# --- Preprocessing ---
def preprocess_dataframe_for_latex(
    df: pd.DataFrame,
    column_transformations: Dict[str, str]
) -> pd.DataFrame:
    """
    Return a new DataFrame with string transformations applied to specified columns,
    and rename columns accordingly.

    Use '*' key to apply the transformation to all object/string columns.
    """
    if not isinstance(df, pd.DataFrame):
        raise TypeError("df must be a pandas DataFrame.")
    if not isinstance(column_transformations, dict):
        raise TypeError("column_transformations must be a dict.")

    result = df.copy()
    transforms = column_transformations.copy()

    # Apply wildcard transformation to all string columns
    if '*' in transforms:
        default_transform = transforms.pop('*')
        for col in result.select_dtypes(include=['object', 'string']).columns:
            transforms.setdefault(col, default_transform)

    # Apply value transformations
    for column, transform in transforms.items():
        if column not in result.columns:
            logger.warning("Column '%s' not found. Skipping.", column)
            continue
        result[column] = result[column].map(
            lambda x: apply_string_transformation(x, transform)
        )

    # Rename columns based on their transformation
    rename_map = {}
    for column, transform in transforms.items():
        new_name = apply_string_transformation(column, transform)
        if new_name != column:
            rename_map[column] = new_name
    result.rename(columns=rename_map, inplace=True)

    # Round numeric columns
    numeric_cols = result.select_dtypes(include='number').columns
    result[numeric_cols] = result[numeric_cols].round(2)
    return result

# --- LaTeX Export ---
def escape_latex(text: str) -> str:
    """
    Escape LaTeX special characters in a string.
    """
    replacements = {
        '&': r'\\&',
        '%': r'\\%',
        '$': r'\\$',
        '#': r'\\#',
        '{': r'\\{',
        '}': r'\\}',
        '~': r'\\textasciitilde{}',
        '^': r'\\textasciicircum{}',
        '\\': r'\\textbackslash{}',
    }
    for old, new in replacements.items():
        text = text.replace(old, new)
    return text


def dataframe_to_latex(
    df: pd.DataFrame,
    caption: str,
    label: str,
    alignment: Optional[str] = None,
    float_format: str = "%.2f",
    na_rep: str = '-',
    escape_text: bool = True,
    escape_header: bool = True,
    index: bool = False,
    header: bool = True,
    bold_header: bool = True,
    use_booktabs: bool = True
) -> str:
    """
    Convert DataFrame to LaTeX table code.
    """
    table_lines = []
    df_work = df
    if index:
        df_work = df_work.reset_index()
        idx_name = df.index.name or 'Index'
        df_work.rename(columns={df_work.columns[0]: idx_name}, inplace=True)

    cols = list(df_work.columns)
    n_cols = len(cols)

    # Determine alignment
    if alignment and len(alignment) == n_cols:
        align = alignment
    else:
        align = 'l' + 'c' * (n_cols - 1)

    # Begin table
    table_lines.append(r"\begin{table}[h!]")
    table_lines.append(r"  \centering")
    table_lines.append(f"  \caption{{{caption}}}")
    table_lines.append(f"  \label{{{label}}}")
    table_lines.append(r"  \begin{tabular}{@{}" + align + r"@{}}")
    top = r"  \toprule" if use_booktabs else r"  \hline"
    mid = r"  \midrule" if use_booktabs else r"  \hline"
    bot = r"  \bottomrule" if use_booktabs else r"  \hline"
    table_lines.extend([top])

    # Header row
    if header:
        header_cells = []
        for col in cols:
            cell = escape_latex(col) if escape_header else col
            if bold_header:
                cell = f"\\textbf{{{cell}}}"
            header_cells.append(cell)
        table_lines.append("  " + " & ".join(header_cells) + r" \\")
        table_lines.append(mid)

    # Data rows
    for _, row in df_work.iterrows():
        row_cells = []
        for val in row:
            if pd.isna(val):
                cell = na_rep
            elif isinstance(val, float):
                cell = float_format % val
                if escape_text:
                    cell = escape_latex(cell)
            else:
                cell = escape_latex(str(val)) if escape_text else str(val)
            row_cells.append(cell)
        table_lines.append("  " + " & ".join(row_cells) + r" \\")

    table_lines.extend([bot, r"  \end{tabular}", r"\end{table}"])
    return "\n".join(table_lines)

# Drop specified columns
def drop_columns(df: pd.DataFrame, cols_to_drop: list) -> pd.DataFrame:
    """Return a new DataFrame without specified columns."""
    return df.drop(columns=cols_to_drop, errors='ignore')

def filter_top_group(
    df: pd.DataFrame,
    group_col: str,
    metric_col: str,
    n_best: int = 3,
    kind: str = 'minor'
) -> pd.DataFrame:
    """
    For each unique group_col value, keep the top n_best rows
    sorted by metric_col. kind='minor' → lowest values, 'major' → highest.
    """
    # decide sort direction for the metric
    ascending = kind.lower().startswith('minor')
    # sort so that best rows appear first within each group
    sorted_df = df.sort_values([group_col, metric_col], ascending=[True, ascending])
    # take top n_best per group without adding extra index levels
    top_df = sorted_df.groupby(group_col, group_keys=False).head(n_best)
    top_df = top_df.sort_values(by=metric_col, ascending=True)
    return top_df.reset_index(drop=True)



if __name__ == "__main__":
    
    transformations = {
        'Well': 'Upper',
        'Category': 'Phrase',
        'strategy': 'Phrase',
        'extractor': 'Upper',
        'fuser': 'Capitalize',
        '*': 'Capitalize'
    }
    
    # 1. Pré-processar o DataFrame
    df_preprocessed = preprocess_dataframe_for_latex(sorted_organized_results['default']['cumulative_metrics'].head(100), transformations)
    # df_preprocessed = preprocess_dataframe_for_latex(sorted_organized_results['default']['aggregated_metrics'].head(100), transformations)
    
    df_preprocessed = drop_columns(df_preprocessed, ['WELL', 'Category'])
    df_preprocessed =  filter_top_group(df_preprocessed, group_col = 'Strategy', metric_col = 'SMAPE_TEST', n_best = 3, kind='minor')

    print("--- DataFrame Pré-processado ---")
    print(df_preprocessed.to_string())
    print("\n" + "="*30 + "\n")

    # 2. Gerar o código LaTeX a partir do DataFrame pré-processado
    latex_code = dataframe_to_latex(
        df_preprocessed,
        caption="Cumulative Forecast Result at Well 15/9-F-14",
        label="F_14_Cumulative",
        float_format="%.2f",
        na_rep="-",
        index=False,
        # Agora não precisamos mais nos preocupar com `_` no escape!
        escape_text=True # Ainda útil para %, $, &, etc.
    )

    # Imprimir o código LaTeX gerado
    print("--- Código LaTeX Gerado ---")
    print(latex_code)
```

```python
import re
from textwrap import dedent
from typing import List, Tuple


def _get_tabular_content(latex_str: str) -> str:
    """Extract content inside the tabular environment."""
    start_tok = r"\\begin{tabular}"
    end_tok = r"\\end{tabular}"
    start = latex_str.find(start_tok)
    if start < 0:
        raise ValueError("Missing \"\\begin{tabular}\"")

    # find end of column spec
    spec_start = latex_str.find("{", start + len(start_tok))
    if spec_start < 0:
        raise ValueError("Malformed begin: missing '{'")
    idx = spec_start + 1
    depth = 1
    while idx < len(latex_str) and depth:
        if latex_str[idx] == '{':
            depth += 1
        elif latex_str[idx] == '}':
            depth -= 1
        idx += 1
    if depth:
        raise ValueError("Unbalanced braces in column spec")

    end = latex_str.find(end_tok, idx)
    if end < 0:
        raise ValueError("Missing \"\\end{tabular}\"")

    return latex_str[idx:end].strip()


def _clean_lines(raw: str) -> List[str]:
    """Filter out rules/comments and strip line endings."""
    lines: List[str] = []
    for line in raw.splitlines():
        text = line.strip()
        if not text or text.startswith(("\\toprule", "\\midrule", "\\bottomrule", "%")):
            continue
        if text.endswith('\\\\'):
            text = text[:-2].strip()
        lines.append(text)
    if not lines:
        raise ValueError("No table lines found")
    return lines


def extract_header_and_rows(latex_str: str) -> Tuple[str, List[str]]:
    """Get header and data rows from a LaTeX table string."""
    content = _get_tabular_content(latex_str)
    lines = _clean_lines(content)
    header, *rows = lines
    if not header:
        raise ValueError("Empty header row")
    return header, rows


def combine_tables_side_by_side(
    table1: str,
    table2: str,
    label1: str,
    label2: str,
    caption: str,
    label: str,
    cols_per: int = 6,
) -> str:
    """Merge two LaTeX tables side by side under one float."""
    hdr1, rows1 = extract_header_and_rows(table1)
    hdr2, rows2 = extract_header_and_rows(table2)

    if hdr1 != hdr2:
        # use first header if they differ
        hdr2 = hdr1

    if len(rows1) != len(rows2):
        raise ValueError(
            f"Row count mismatch: {len(rows1)} != {len(rows2)}"
        )

    spec = '@{}lllccc@{\quad}lllccc@{}'
    head_line = (
        f"\\multicolumn{{{cols_per}}}{{c}}{{\\textbf{{{label1}}}}} & "
        f"\\multicolumn{{{cols_per}}}{{c}}{{\\textbf{{{label2}}}}} \\\\"
    )
    col_line = f"{hdr1} & {hdr2} \\\\"

    data_lines = []
    for r1, r2 in zip(rows1, rows2):
        data_lines.append(f"  {r1} & {r2} \\\\")
    data_block = "\n".join(data_lines)

    table = dedent(f"""
    \begin{{table}}[h!]
      \centering
      \caption{{{caption}}}
      \label{{{label}}}
      \resizebox{{\textwidth}}{{!}}{{%
        \begin{{tabular}}{{{spec}}}
          \toprule
          {head_line}
          \midrule
          {col_line}
          \midrule
{data_block}
          \bottomrule
        \end{{tabular}} %
      }}
    \end{{table}}
    """
    )
    return table.strip()


# --- Example Usage ---
if __name__ == "__main__":


    # Combine the tables
    combined_table = combine_latex_tables_side_by_side(
        latex_table1_str=latex_code,
        latex_table2_str=latex_code,
        well1_name="15/9-F-14",
        well2_name="15/9-F-12",
        combined_caption="Side-by-Side Cumulative Forecast Results for Wells F-14 and F-12",
        combined_label="F-14_F-12_Cumulative_Combined",
        num_cols_per_table=6 # Strategy, EXTRACTOR, Fuser, R², MAE, SMAPE
    )

    print("\n--- Combined LaTeX Table ---")
    print(combined_table)
```
